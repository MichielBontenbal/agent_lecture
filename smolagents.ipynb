### When to use agents

Why would you use agents? Well, LLM's are limited in their knowledge.

One scenario is where you have multiple sub-questions or dependencies in your task.
A good example is this: "I am in Amsterdam. I would like to spend this weekend in Paris, but only if there is no rain and a single trip train ticket is cheaper than 100 euro."

## Planning a trip to Paris 

```python
#setting the prompt

prompt = "I would like to go this weekend to Paris, but only if there is no rain and the train ticket from Amsterdam to Paris is cheaper than 100 euro."
```

```python
# run the code
from smolagents import CodeAgent, DuckDuckGoSearchTool, InferenceClientModel

agent = CodeAgent(tools=[DuckDuckGoSearchTool()], model=InferenceClientModel(token=hf_token))

agent.run(prompt)
```
